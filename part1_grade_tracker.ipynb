```python
##roll is converted from a string to an integer.
##marks_str is split on ", " and each element is converted to an integer, producing a marks list.
clear_list=[]
for std in raw_students: ##Looping through raw_students
    clear_name = std["name"].strip().title() ##whitespace and is converted to Title Case.
    clear_roll=int(std["roll"])
    clear_marks=[int(m) for m in std["marks_str"].split(",")]
    is_valid= clear_name.replace(" ", "").isalpha()
    validation_msg = "✓ Valid name" if is_valid else "✗ Invalid name" # Name validation message
    cleared_std = {
    "name": clear_name,
    "roll": clear_roll,
    "marks": clear_marks #Cleaned dictionary to use in later parts
}  
    clear_list.append(cleared_std)
    
    #Printing formatted profile card 
    print(f"============================================")
    print(f"Student : {clear_name} ({validation_msg})")
    print(f"Roll no : {clear_roll}")
    print(f"Marks : {clear_marks}")
    print(f"============================================")

for c in clear_list:
    if c["roll"] == 103:
        print(f"Lower Case: {c["name"].lower()}")
        print(f"Upper Case:{c["name"].upper()}")


student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
L=len(subjects)
for s in range(L):
    m = marks[s]
    if m>=90:
        grade = "A+"
    if m>=80 and m<90:
        grade="A"
    if m>=70 and m<80:
        grade="B"
    if m>=60 and m<70:
        grade="C"
    if m<=59:
        grade="F"
    print(f"{subjects[s]:100}: {m} {grade}")
```

```python
# Calculations of total, average, highest and lowest marks
total_marks=sum(marks)
avg_marks= round(total_marks / len(marks), 2)
high_idx = marks.index(max(marks))
low_idx = marks.index(min(marks))

print(f"\nTotal Marks: {total_marks}")
print(f"Average Marks: {avg_marks}")
print(f"Highest Score : {subjects[high_idx]} {marks[high_idx]}")
print(f"Lowest Score : {subjects[low_idx]} {marks[low_idx]}")


total_marks= 0
subject_count=0
while True:
    sub_input = input("Enter subject name: ").strip()
    if sub_input.lower() == "done":
        break
    marks_input = input(f"Enter marks for {sub_input}:")
  # Check if input is a number (basics only)
    if marks_input.isdigit():
        marks = int(marks_input)
        # Check if the number is in the valid range
        if marks >= 0 and marks <= 100:
            total_marks = total_marks + marks
            subject_count = subject_count + 1
        else:
            print("Warning: Marks must be between 0 and 100.")
    else:
        print("Warning: Please enter a whole number.")
        # Final calculation
if subject_count > 0:
    average = total_marks / subject_count
    print("\n--- Results ---")
    print("Subjects added:", subject_count)
    print("Average marks:", average)
else:
    print("No subjects were entered.")

```

```python
## Task 4: String Manipulation Utility
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
#Stripping and lowercasing 
clean_essay = essay.strip().lower()
print(f"Step 1 (Cleaned): {clean_essay}")
# Convert to Title Case
print(f"Step 2 (Title Case): {clean_essay.title()}")
# Convert "python" (Case-insensitive)
python_count = clean_essay.count("python")
print(f"Step 3 (Count): {python_count}")
#Replacing with emoji
emoji_essay = clean_essay.replace("python", "Python 🐍")
print(f"Step 4 (replacement): {emoji_essay}")
#Split into sentences
sentences = clean_essay.split(". ")
print(f"Step 5 (Split List): {sentences}")
# Numbered sentences with punctuation check
print("\nStep 6 (Numbered Sentences):")
for i, sentence in enumerate(sentences, 1):
    #Stripping any leftover whitespace
    s = sentence.strip()
    #Skip empty strings
    if not s:
        continue
    #Check if sentence ends with a period, if not add it back
    if not s.endswith("."):
        s += "."
    #Capitlize first letter of sentences
    print(f"{i}, {s.capitalize()}")

```
